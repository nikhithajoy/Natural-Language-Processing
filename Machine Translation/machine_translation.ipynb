We import the pipeline class from transformers to create a translation pipeline.

```python
from transformers import pipeline
```

We define a translation pipeline ("translation_en_to_fr") for translating English (en) to French (fr).

```python
# Define the translation pipeline using the 'translation_xx_to_yy' task
translator = pipeline("translation_en_to_fr")
```

```python
# Input text to translate
input_text = "Hello, how are you?"
```

```python
# Translate the input text from English to French
translated_text = translator(input_text)
```

```python
# Display the translated text
print("Input Text (English):", input_text)
print("Translated Text (French):", translated_text[0]['translation_text'])
```
